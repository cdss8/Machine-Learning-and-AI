```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/gdrive/My Drive/AI-ML/day.csv')
data.head()
```

```python
selectedData = data.drop(['instant', 'casual', 'registered', 'atemp', 'dteday'], axis=1)
selectedData
```

```python
x=selectedData.drop(['cnt'], axis=1)
y=selectedData['cnt']
```

```python
#train&testDataSelection

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
print('initial =',x.shape,'train =',x_train.shape,'test =', x_test.shape)
```

```python
#Calculating the LinearRegresion

from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train,y_train)
print(LR.coef_)
print(LR.intercept_)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

Y_predLR= LR.predict(x_test)

dataScores = pd.DataFrame({'R2-Score': round(r2_score(y_test,Y_predLR),3),
'MSE': round(mean_squared_error(y_test,Y_predLR),3),
'MAE': round(mean_absolute_error(y_test,Y_predLR),3)}, index = ['LinearRegression'])

dataScores
```

```python
#LinearR Data visualitation

Y_predL = Y_predLR
plt.figure(figsize=(15,5))
plt.plot(y_test.reset_index(drop=True),color='g')
plt.plot(Y_predL,color='r')
```

```python
#Calculating the PolinomialRegresion (exponent 2)

from sklearn.preprocessing import PolynomialFeatures

PRsquare = PolynomialFeatures(degree=2)

x_train_square = PRsquare.fit_transform(x_train)
x_test_square = PRsquare.fit_transform(x_test)
x_train_square.shape
```

```python
PRsquare_model = LinearRegression()
PRsquare_model.fit(x_train_square, y_train)

Y_predPRsquare_model = PRsquare_model.predict(x_test_square)

dataScores = dataScores.append(pd.DataFrame({'R2-Score':
round(r2_score(y_test,Y_predPRsquare_model),3),
'MSE': round(mean_squared_error(y_test,Y_predPRsquare_model),3),
'MAE': round(mean_absolute_error(y_test,Y_predPRsquare_model),3)}, index = ['QuadraticRegression']))

dataScores

#I have ran the code many times, is that i got duplicates
```

```python
#SqureR Data Visualitation (Polinomial)

Y_predS = Y_predPRsquare_model
plt.figure(figsize=(15,5))
plt.plot(y_test.reset_index(drop=True),color='b')
plt.plot(Y_predS,color='r')
```

```python
#Calculating the PolinomialRegresion (exponent 3)

PRcube = PolynomialFeatures(degree=3)

x_train_cube = PRcube.fit_transform(x_train)
x_test_cube = PRcube.fit_transform(x_test)

PRcube_model = LinearRegression()
PRcube_model.fit(x_train_cube,y_train)

y_pred_cubic_model = PRcube_model.predict(x_test_cube)

dataScores = dataScores.append(pd.DataFrame({'R2-Score':
round(r2_score(y_test,y_pred_cubic_model),3),
'MSE': round(mean_squared_error(y_test,y_pred_cubic_model),3),
'MAE': round(mean_absolute_error(y_test,y_pred_cubic_model),3)}, index =['CubicRegression']))

dataScores
```

```python
#CubicR Data visualization (Polinomial)

Y_predC = y_pred_cubic_model
plt.figure(figsize=(15,5))
plt.plot(y_test.reset_index(drop=True),color='b')
plt.plot(Y_predC,color='g')
```

```python
#Calculating the ForestRegresion (Regresion based in a Forest)

from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(n_estimators=300, max_features='sqrt', bootstrap=False, random_state=1)
forest_model.fit(x_train,y_train)

y_predFM = forest_model.predict(x_test)

dataScores = dataScores.append(pd.DataFrame({'R2-Score':round(r2_score(y_test,y_predFM),3),
'MSE': round(mean_squared_error(y_test,y_predFM),3),
'MAE': round(mean_absolute_error(y_test,y_predFM),3)}, index = ['RandomForestRegression']))

#Calculating the ForestRegresion (Regresion based in a Forest)dataScores
```

```python
#Forest Data visualization

Y_predF = y_predFM
plt.figure(figsize=(15,5))
plt.plot(y_test.reset_index(drop=True),color='m')
plt.plot(Y_predF,color='g')
```

```python
#Calculating as optional the AdaBoostRegresion

from sklearn.ensemble import AdaBoostRegressor

adaBoost_model = AdaBoostRegressor(n_estimators=50, loss='linear', random_state=1,learning_rate = 3)
adaBoost_model.fit(x_train,y_train)

y_predAM = adaBoost_model.predict(x_test)

dataScores = dataScores.append(pd.DataFrame({'R2-Score': round(r2_score(y_test,y_predAM),3), 
'MSE': round(mean_squared_error(y_test,y_predAM),3),
'MAE': round(mean_absolute_error(y_test,y_predAM),3)}, index =['AdaBoostRegression']))

dataScores
```

```python
#Forest Data visualization

Y_predA = y_predAM
plt.figure(figsize=(15,5))
plt.plot(y_test.reset_index(drop=True),color='c')
plt.plot(Y_predA,color='r')
```
